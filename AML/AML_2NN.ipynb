# SCRATCH IMPLEMENTATION --

```python
import numpy as np

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [0]])
input_units = X.shape[1]
hidden_units = 4
output_units = Y.shape[1]

print(f"Input units: {input_units}")
print(f"Hidden units: {hidden_units}")
print(f"Output units: {output_units}")

np.random.seed(42)
W1 = np.random.randn(input_units, hidden_units) * 0.01

W2 = np.random.randn(hidden_units, output_units) * 0.01

print(f"\nShape of W1: {W1.shape}")
print(f"Shape of W2: {W2.shape}")

b1 = np.zeros((1, hidden_units))

b2 = np.zeros((1, output_units))

print(f"Shape of b1: {b1.shape}")
print(f"Shape of b2: {b2.shape}")

print('\nNeural network architecture defined and weights/biases initialized.')
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    return x * (1 - x)
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

print("Activation functions (sigmoid) and loss function (MSE) defined.")
```

```python
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)

    return Z1, A1, Z2, A2

print("Forward propagation function defined.")
```

```python
def backward_propagation(X, Y, A1, A2, W2):
    m = X.shape[0]
    dZ2 = A2 - Y

    dW2 = np.dot(A1.T, dZ2) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    dZ1 = np.dot(dZ2, W2.T) * sigmoid_derivative(A1)

    dW1 = np.dot(X.T, dZ1) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    return dW1, db1, dW2, db2

print("Backward propagation function defined.")
```

```python
epochs = 10000
learning_rate = 0.1

losses = []

for i in range(epochs):
    Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)

    loss = mean_squared_error(Y, A2)
    losses.append(loss)

    dW1, db1, dW2, db2 = backward_propagation(X, Y, A1, A2, W2)

    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2

    if i % 1000 == 0:
        print(f"Epoch {i}, Loss: {loss:.4f}")

print("\nTraining complete!")
```

```python
def backward_propagation(X, Y, A1, A2, W2):
    m = X.shape[0]

    dZ2 = (A2 - Y) * sigmoid_derivative(A2)

    dW2 = np.dot(A1.T, dZ2) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    dZ1 = np.dot(dZ2, W2.T) * sigmoid_derivative(A1)

    dW1 = np.dot(X.T, dZ1) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    return dW1, db1, dW2, db2

print("Corrected backward propagation function defined.")
```

```python
epochs = 10000
learning_rate = 1.0

losses = []

np.random.seed(42)
W1 = np.random.randn(input_units, hidden_units) * 0.01
W2 = np.random.randn(hidden_units, output_units) * 0.01
b1 = np.zeros((1, hidden_units))
b2 = np.zeros((1, output_units))

print("Re-initializing weights and biases and increasing learning rate for re-training.")

for i in range(epochs):
    Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)

    loss = mean_squared_error(Y, A2)
    losses.append(loss)

    dW1, db1, dW2, db2 = backward_propagation(X, Y, A1, A2, W2)

    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2

    if i % 1000 == 0:
        print(f"Epoch {i}, Loss: {loss:.4f}")

print("\nTraining complete with updated learning rate!")
```

```python
print('Evaluating the trained neural network...')

_, _, _, A2_pred = forward_propagation(X, W1, b1, W2, b2)

predictions = (A2_pred >= 0.5).astype(int)

print('\nInput, Target, Prediction:')
for i in range(len(X)):
    print(f"Input: {X[i]}, Target: {Y[i][0]}, Predicted: {predictions[i][0]}")
```

```python
import matplotlib.pyplot as plt

print('Generating decision boundary plot...')

x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5
h = 0.01

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

grid_points = np.c_[xx.ravel(), yy.ravel()]

_, _, _, Z = forward_propagation(grid_points, W1, b1, W2, b2)

Z = (Z >= 0.5).astype(int)

Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)

plt.scatter(X[Y[:,0] == 0, 0], X[Y[:,0] == 0, 1], color='red', marker='o', s=100, label='Class 0 (XOR 0)')
plt.scatter(X[Y[:,0] == 1, 0], X[Y[:,0] == 1, 1], color='blue', marker='x', s=100, label='Class 1 (XOR 1)')

plt.title('XOR Decision Boundary')
plt.xlabel('Input Feature 1')
plt.ylabel('Input Feature 2')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.legend()
plt.grid(True)

plt.show()

print("Decision boundary plot generated.")
```

# SKLEARN IMPLEMENTATION --

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.exceptions import ConvergenceWarning
import warnings

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 0])

print("XOR dataset defined.")

warnings.filterwarnings("ignore", category=ConvergenceWarning)

mlp = MLPClassifier(hidden_layer_sizes=(4,), activation='logistic',
                    solver='lbfgs', max_iter=10000, random_state=42)

print("MLPClassifier initialized with one hidden layer of 4 neurons, sigmoid activation, and L-BFGS solver.")
```

```python
print("Training the MLPClassifier...")
mlp.fit(X, Y)
print("Training complete.")

predictions = mlp.predict(X)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y, predictions)

print("\nMLPClassifier Predictions for XOR dataset:")
for i in range(len(X)):
    print(f"Input: {X[i]}, Target: {Y[i]}, Predicted: {predictions[i]}")

print(f"\nAccuracy: {accuracy*100:.2f}%")
```

```python
import matplotlib.pyplot as plt

print('Generating decision boundary plot for Scikit-learn MLPClassifier...')

x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5
h = 0.01

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = mlp.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)

plt.scatter(X[Y == 0, 0], X[Y == 0, 1], color='red', marker='o', s=100, label='Class 0 (XOR 0)')
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='blue', marker='x', s=100, label='Class 1 (XOR 1)')

plt.title('XOR Decision Boundary - Scikit-learn MLPClassifier')
plt.xlabel('Input Feature 1')
plt.ylabel('Input Feature 2')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.legend()
plt.grid(True)

plt.show()

print("Decision boundary plot generated for Scikit-learn MLPClassifier.")
```
