```python
cgpa1=[9.2,6.2,8.6,8.2,6.7,5.4,6,8]
cgpa2=[7,7.7,6.7,9,8,5.6,8,6]
cgpa3=[7.2,6.7,7.7,7,6,6.6,7,5.4]
cgpa4=[7.6,8.8,8.7,9,8,7.6,9,9.4]
sem=[1,2,3,4,5,6,7,8]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(sem, cgpa1, marker='o', label='CGPA 1')

plt.grid(False)
plt.show()
```

```python
plt.plot(sem, cgpa1, marker='o')
plt.ylim(0, 10)
plt.show()
```

```python
plt.plot(sem, cgpa1, marker='o')
plt.ylim(0, 10)
plt.xlabel("Semester")
plt.ylabel("CGPA")
plt.title("CGPA vs Semester (cgpa1)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

bar_width = 0.2

r1 = np.array(sem) - 1.5*bar_width
r2 = np.array(sem) - 0.5*bar_width
r3 = np.array(sem) + 0.5*bar_width
r4 = np.array(sem) + 1.5*bar_width

plt.bar(r1, cgpa1, width=bar_width, label='cgpa1')
plt.bar(r2, cgpa2, width=bar_width, label='cgpa2')
plt.bar(r3, cgpa3, width=bar_width, label='cgpa3')
plt.bar(r4, cgpa4, width=bar_width, label='cgpa4')


plt.xlabel("Semester")
plt.ylabel("CGPA")
plt.title("Semester VS CGPA")
plt.xticks(sem)
plt.ylim(0, 10)

plt.show()
```

```python
plt.plot(sem, cgpa1, label="L1")
plt.plot(sem, cgpa2, label="L2")
plt.plot(sem, cgpa3, label="L3")

plt.xlabel("Semester")
plt.ylabel("CGPA")
plt.title("Semester VS CGPA")
plt.ylim(0, 10)
plt.legend()

plt.show()
```

```python

cgpa1 = [9.2, 6.2, 8.6, 8.2, 6.7, 5.4, 6, 8]
cgpa2 = [7, 7.7, 6.7, 9.8, 5.6, 8, 6, 7]
sem   = [1, 2, 3, 4, 5, 6, 7, 8]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

ax1.plot(sem, cgpa1)
ax1.set_ylim(5, 10)

ax2.plot(sem, cgpa2)
ax2.set_ylim(5, 10)

plt.show()
```

```python



fig, axs = plt.subplots(2, 2, figsize=(10,6))


axs[0,0].plot(sem, cgpa1)
axs[0,0].set_ylim(5, 10)


axs[0,1].plot(sem, cgpa2)
axs[0,1].set_ylim(5, 10)


axs[1,0].bar(sem, cgpa3)
axs[1,0].set_ylim(0, 10)


axs[1,1].bar(sem, cgpa4)
axs[1,1].set_ylim(0, 10)


plt.tight_layout()
plt.show()
```

```python


fig, axs = plt.subplots(2, 2, figsize=(10,6))

axs[0,0].plot(sem, cgpa1)
axs[0,0].set_title("Year VS Price")
axs[0,0].set_xlabel("Year")
axs[0,0].set_ylabel("Price")
axs[0,0].set_ylim(5, 10)

axs[0,1].plot(sem, cgpa2)
axs[0,1].set_title("Year VS Price")
axs[0,1].set_xlabel("Year")
axs[0,1].set_ylabel("Price")
axs[0,1].set_ylim(5, 10)

axs[1,0].bar(sem, cgpa3)
axs[1,0].set_title("Year VS Price")
axs[1,0].set_xlabel("Year")
axs[1,0].set_ylabel("Price")
axs[1,0].set_ylim(0, 10)

axs[1,1].bar(sem, cgpa4)
axs[1,1].set_title("Year VS Price")
axs[1,1].set_xlabel("Year")
axs[1,1].set_ylabel("Price")
axs[1,1].set_ylim(0, 10)

plt.tight_layout()
plt.show()
```
