# SLR (FAKE DATASET)

```python
import numpy as np

np.random.seed(0)
X_simple = np.linspace(0, 10, 100).reshape(-1, 1)
y_simple = 2 * X_simple + 5 + np.random.randn(100, 1) * 2

X_simple_b = np.c_[np.ones((100, 1)), X_simple]

theta_simple = np.linalg.inv(X_simple_b.T.dot(X_simple_b)).dot(X_simple_b.T).dot(y_simple)

print("Simple Linear Regression Coefficients (theta):")
print(theta_simple)
```

# MLR (FAKE DATASET)

```python
import numpy as np

np.random.seed(0)
X_multiple = np.random.rand(100, 2) * 10
y_multiple = 3 * X_multiple[:, 0].reshape(-1, 1) + 4 * X_multiple[:, 1].reshape(-1, 1) + 10 + np.random.randn(100, 1) * 5

X_multiple_b = np.c_[np.ones((100, 1)), X_multiple]

theta_multiple = np.linalg.inv(X_multiple_b.T.dot(X_multiple_b)).dot(X_multiple_b.T).dot(y_multiple)

print("\nMultiple Linear Regression Coefficients (theta):")
print(theta_multiple)
```
