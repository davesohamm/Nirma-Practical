

The `pd.concat()` function is used to combine DataFrames either row-wise (`axis=0`) or column-wise (`axis=1`).


```python

dataa = pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]})
datab = pd.DataFrame({'col1': [5, 6], 'col2': [7, 8]})

dataa = pd.DataFrame({'colA': [1, 2], 'colB': [3, 4]})
datab = pd.DataFrame({'colA': [5, 6], 'colB': [7, 8]})
display(dataa)
display(datab)

datac = pd.concat([dataa, datab], axis=0)

print("Resulting DataFrame:")
display(datac)
```



Use `pd.concat()` with `axis=1` to combine DataFrames column-wise.

```python

datac_colwise = pd.concat([dataa, datab], axis=1)

display(datac_colwise)
```



Merging DataFrames is similar to SQL joins. You combine DataFrames based on common columns (keys).


```python
EMP = pd.DataFrame({'EID': [1, 2, 3, 4], 'Name': ['Soham', 'Dev', 'Abhi', 'Vraj'], 'CID': [101, 102, 101, 103]})
COMP = pd.DataFrame({'CID': [101, 102, 104], 'CompanyName': ['Apple Inc.', 'Google Corp.', 'Microsoft Ltd.']})

display(EMP)
display(COMP)

print("\nInner join of EMP and COMP on 'CID':")
inner_merged_df = pd.merge(EMP, COMP, on='CID')
display(inner_merged_df)


print("\nLeft outer join of EMP and COMP on 'CID':")
left_merged_df = pd.merge(EMP, COMP, on='CID', how='left')
display(left_merged_df)


print("\nRight outer join of EMP and COMP on 'CID':")
right_merged_df = pd.merge(EMP, COMP, on='CID', how='right')
display(right_merged_df)


print("\nFull outer join of EMP and COMP on 'CID':")
outer_merged_df = pd.merge(EMP, COMP, on='CID', how='outer')
display(outer_merged_df)
```
