1. Write a program which scans marks of 200 students in P & S. Assuming that the students are to be graded on 5 grade-scale (A, B, C, D, and F), display marks and grade of each student. Fit Gaussian distribution to the data. Repeat the exercise if students are to be graded on 7 grade-scale (A, B, C+, C, C-, D, F)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

student_marks = np.random.randint(0, 101, 200)
student_df = pd.DataFrame({'Roll Number': range(1, len(student_marks) + 1),
                           'Science Marks': student_marks})

student_df.to_csv('student_marks.csv', index=False)

display(student_df.head())
```

```python
mean_marks = student_marks.mean()
std_dev_marks = student_marks.std()

print(f"Mean of student marks: {mean_marks:.2f}")
print(f"Standard deviation of student marks: {std_dev_marks:.2f}")
```

```python
mean_marks = student_marks.mean()
std_dev_marks = student_marks.std()

z_score_grade_scale_5 = {
    'A': 1.5,
    'B': 0.5,
    'C': -0.5,
    'D': -1.5,
    'F': -np.inf
}

z_score_grade_scale_7 = {
    'A': 1.5,
    'B': 1.0,
    'C+': 0.5,
    'C': 0,
    'C-': -0.5,
    'D': -1.0,
    'F': -np.inf
}

print("Relative Grade Boundaries (5-scale - Z-score):")
for grade, boundary in z_score_grade_scale_5.items():
    print(f"{grade}: > {boundary}")

print("\nRelative Grade Boundaries (7-scale - Z-score):")
for grade, boundary in z_score_grade_scale_7.items():
    print(f"{grade}: > {boundary}")
```

```python
relative_grades_5scale = []
for mark in student_marks:

    z_score = (mark - mean_marks) / std_dev_marks


    if z_score > z_score_grade_scale_5['A']:
        relative_grades_5scale.append('A')
    elif z_score > z_score_grade_scale_5['B']:
        relative_grades_5scale.append('B')
    elif z_score > z_score_grade_scale_5['C']:
        relative_grades_5scale.append('C')
    elif z_score > z_score_grade_scale_5['D']:
        relative_grades_5scale.append('D')
    else:
        relative_grades_5scale.append('F')

student_df['Relative Grade (5-scale)'] = relative_grades_5scale
```

```python
for index, row in student_df.iterrows():
    print(f"Student {row['Roll Number']}: Mark = {row['Science Marks']}, Relative Grade (5-scale) = {row['Relative Grade (5-scale)']}")
```

```python
relative_grades_7scale = []
for mark in student_marks:
    z_score = (mark - mean_marks) / std_dev_marks
    if z_score > z_score_grade_scale_7['A']:
        relative_grades_7scale.append('A')
    elif z_score > z_score_grade_scale_7['B']:
        relative_grades_7scale.append('B')
    elif z_score > z_score_grade_scale_7['C+']:
        relative_grades_7scale.append('C+')
    elif z_score > z_score_grade_scale_7['C']:
        relative_grades_7scale.append('C')
    elif z_score > z_score_grade_scale_7['C-']:
        relative_grades_7scale.append('C-')
    elif z_score > z_score_grade_scale_7['D']:
        relative_grades_7scale.append('D')
    else:
        relative_grades_7scale.append('F')

student_df['Relative Grade (7-scale)'] = relative_grades_7scale

for index, row in student_df.iterrows():
    print(f"Student {row['Roll Number']}: Mark = {row['Science Marks']}, Relative Grade (7-scale) = {row['Relative Grade (7-scale)']}")
```

```python
relative_grade_counts_5scale = student_df['Relative Grade (5-scale)'].value_counts().sort_index()
print("Relative Grade Distribution (5-scale):")
print(relative_grade_counts_5scale)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=relative_grade_counts_5scale.index, y=relative_grade_counts_5scale.values, palette='viridis', hue=relative_grade_counts_5scale.index, legend=False)
plt.title('Relative Grade Distribution (5-grade scale)')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.show()
```

```python
relative_grade_counts_7scale = student_df['Relative Grade (7-scale)'].value_counts().sort_index()
print("Relative Grade Distribution (7-scale):")
print(relative_grade_counts_7scale)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=relative_grade_counts_7scale.index, y=relative_grade_counts_7scale.values, palette='viridis', hue=relative_grade_counts_7scale.index, legend=False)
plt.title('Relative Grade Distribution (7-grade scale)')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.show()
```

2. Write a program which scans value of k where k indicates number of mutually exclusive and exhaustive events (E1, E2, , Ek). Assume any another event B. Implement Bayes theorem assuming these k events and event B to estimate the probability of each of the k events assuming that the event B has already occurred (i.e. P(E1|B), P(E2|B), , P(Ek|B)). Your program should also scan (i) probability of each of the k (or k -1) events and (ii) probability of occurrence of event B assuming that each of the k events has occurred (i.e. P(B|E1), P(B|E2),  P(B|Ek)).

```python
while True:
    try:
        k = int(input("Enter the number of mutually exclusive and exhaustive events (k): "))
        if k > 0:
            break
        else:
            print("Please enter a positive integer.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

print(f"Number of events (k): {k}")
```

```python
prior_probabilities = np.random.rand(k)
prior_probabilities /= prior_probabilities.sum()

print(f"\nGenerated Prior Probabilities for {k} events:")
for i, prob in enumerate(prior_probabilities):
    print(f"P(E{i+1}): {prob:.4f}")

print(f"\nSum of Prior Probabilities: {prior_probabilities.sum():.4f}")
```

```python

conditional_probabilities_b_given_ei = np.random.rand(k)

print("\nGenerated Conditional Probabilities P(B|Ei):")
for i, prob in enumerate(conditional_probabilities_b_given_ei):
    print(f"P(B|E{i+1}): {prob:.4f}")
```

```python
total_prob_b = 0
for i in range(k):
    total_prob_b += conditional_probabilities_b_given_ei[i] * prior_probabilities[i]

print(f"\nTotal Probability of event B (P(B)): {total_prob_b:.4f}")
```

```python
posterior_probabilities_ei_given_b = []
for i in range(k):
    posterior_prob = (conditional_probabilities_b_given_ei[i] * prior_probabilities[i]) / total_prob_b
    posterior_probabilities_ei_given_b.append(posterior_prob)

print("\nPosterior Probabilities P(Ei|B):")
for i, prob in enumerate(posterior_probabilities_ei_given_b):
    print(f"P(E{i+1}|B): {prob:.4f}")
```
