```python
import random

random_numbers = [random.randint(0, 100) for _ in range(1000)]
tot_mean = sum(random_numbers) / len(random_numbers)
print(tot_mean)
tot_std = (sum([(x - tot_mean) ** 2 for x in random_numbers]) / len(random_numbers)) ** 0.5
print(tot_std)
```

```python
import numpy as np

means = []
stds = []

for _ in range(500):
  start_index = random.randint(0, len(random_numbers) - 100)
  batch = random_numbers[start_index : start_index + 100]
  batch_mean = np.mean(batch)
  batch_std = np.std(batch)
  means.append(batch_mean)
  stds.append(batch_std)

print(means)
print(stds)
sample_mean = np.mean(means)
sample_std = np.mean(stds)
```

```python
import matplotlib.pyplot as plt

plt.hist(means, bins=20)
plt.axvline(tot_mean, color='red', linestyle='dashed', linewidth=1, label=f'Population Mean: {tot_mean:.2f}')
plt.axvline(sample_mean, color='black', linestyle='dashed', linewidth=1, label=f'Average of Sample Means: {tot_mean:.2f}')
plt.title("Distribution of Batch Means")
plt.xlabel("Batch Mean")
plt.legend()
plt.show()

plt.hist(stds, bins=20)
plt.axvline(tot_std, color='red', linestyle='dashed', linewidth=1, label=f'Population Std Dev: {tot_std:.2f}')
plt.axvline(sample_std, color='black', linestyle='dashed', linewidth=1, label=f'Average of Sample Std Devs: {sample_std:.2f}')
plt.title("Distribution of Batch Standard Deviations")
plt.xlabel("Batch Standard Deviation")
plt.legend()
plt.show()
```
