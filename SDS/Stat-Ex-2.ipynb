```python
import plotly.graph_objects as go
from statistics import mean

# Data
data1 = [44, 46, 48, 45, 47]
data2 = [34, 46, 59, 39, 52]
x1 = list(range(1, len(data1) + 1))
x2 = list(range(1, len(data2) + 1))

mean1 = mean(data1)
mean2 = mean(data2)

# Create Figure
fig = go.Figure()

# Dataset 1
fig.add_trace(go.Scatter(
    x=x1,
    y=data1,
    mode='lines+markers',
    name='Dataset 1',
    line=dict(color='blue'),
    marker=dict(size=10),
    hoverinfo='x+y'
))

# Mean line for Dataset 1
fig.add_trace(go.Scatter(
    x=[x1[0], x1[-1]],
    y=[mean1, mean1],
    mode='lines',
    name='Mean (Dataset 1)',
    line=dict(color='blue', dash='dash')
))

# Dataset 2
fig.add_trace(go.Scatter(
    x=[i + 0.1 for i in x2],  # slight shift for clarity
    y=data2,
    mode='lines+markers',
    name='Dataset 2',
    line=dict(color='orange'),
    marker=dict(size=10),
    hoverinfo='x+y'
))

# Mean line for Dataset 2
fig.add_trace(go.Scatter(
    x=[x2[0] + 0.1, x2[-1] + 0.1],
    y=[mean2, mean2],
    mode='lines',
    name='Mean (Dataset 2)',
    line=dict(color='orange', dash='dash')
))

# Layout
fig.update_layout(
    title='Dispersion Comparison: Dataset 1 (Clustered) vs Dataset 2 (Dispersed)',
    xaxis_title='Index',
    yaxis_title='Value',
    yaxis=dict(range=[30, 65]),
    hovermode='x unified',
    template='plotly_white'
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = ['Sex', 'Embarked', 'Pclass', '2urvived']

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df_titanic)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
print("Interpretation of Bar Charts:")
print("\nDistribution of Sex:")
print("The bar chart shows that there are significantly more individuals with 'Sex' value 0 than 'Sex' value 1. Assuming 0 represents male and 1 represents female, this indicates a higher number of male passengers on the Titanic.")

print("\nDistribution of Embarked:")
print("The bar chart displays the distribution of 'Embarked' values. Value 2 has the highest count, followed by value 0, and then value 1. Without the mapping of these values to port names, it indicates the relative frequency of passengers embarking from different locations.")

print("\nDistribution of Pclass:")
print("The bar chart shows the distribution of passenger classes ('Pclass'). Class 3 has the highest number of passengers, followed by Class 1, and then Class 2. This suggests that the majority of passengers were in the third class.")

print("\nDistribution of 2urvived:")
print("The bar chart for '2urvived' (assuming this represents survival status, likely 0 for not survived and 1 for survived) indicates that a much larger number of passengers did not survive (value 0) compared to those who survived (value 1).")
```

```python
categorical_cols = ['Sex', 'Embarked', 'Pclass', '2urvived']

for col in categorical_cols:
    value_counts = df_titanic[col].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    plt.title(f'Distribution of {col}')
    plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
print("Interpretation of Pie Charts:")
print("\nDistribution of Sex:")
print("The pie chart for 'Sex' shows that approximately 64.4% of the passengers are of one sex (represented by 0), while 35.6% are of the other sex (represented by 1).")

print("\nDistribution of Embarked:")
print("The 'Embarked' pie chart indicates the proportion of passengers from different embarkation points. Approximately 69.9% embarked from location 2, 20.7% from location 0, and 9.4% from location 1.")

print("\nDistribution of Pclass:")
print("The 'Pclass' pie chart illustrates the distribution of passenger classes. The largest proportion of passengers (54.2%) are in Class 3, followed by Class 1 (24.7%) and then Class 2 (21.2%).")

print("\nDistribution of 2urvived:")
print("The pie chart for '2urvived' (survival status) shows that approximately 73.9% of the passengers did not survive (represented by 0), while 26.1% survived (represented by 1).")
```

```python
categorical_cols = ['Sex', 'Embarked', 'Pclass', '2urvived']

for col in categorical_cols:
    value_counts = df_titanic[col].value_counts()
    plt.figure(figsize=(6, 4))
    sns.barplot(x=value_counts.index, y=value_counts.values)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

print("Interpretation of Column Charts:")
print("\nDistribution of Sex:")
print("The column chart for 'Sex' clearly shows that the count for category 0 is significantly higher than for category 1. This suggests there were more individuals in category 0 in the dataset.")

print("\nDistribution of Embarked:")
print("The 'Embarked' column chart indicates that category 2 has the highest count, followed by category 0, and then category 1. The counts for categories 0 and 1 are considerably lower than category 2.")

print("\nDistribution of Pclass:")
print("The column chart for 'Pclass' shows that category 3 has the highest number of entries, followed by category 1, and then category 2. This means that category 3 is the most frequent passenger class in the dataset.")

print("\nDistribution of 2urvived:")
print("The '2urvived' column chart shows a stark difference between the two categories. Category 0 has a much higher count than category 1. This indicates that there are significantly more individuals in category 0 compared to category 1.")
```

```python
# Relevant categorical columns for stacked bar charts
# Good combinations are those where one column's distribution is interesting in relation to another.
# 'Pclass' and '2urvived' (Survival by Class)
# 'Sex' and '2urvived' (Survival by Sex)
# 'Embarked' and '2urvived' (Survival by Embarkation Point)

categorical_pairs = [
    ('Pclass', '2urvived'),
    ('Sex', '2urvived'),
    ('Embarked', '2urvived')
]

for col1, col2 in categorical_pairs:
    # Group data and unstack for plotting
    grouped_data = df_titanic.groupby(col1)[col2].value_counts().unstack().fillna(0)

    # Create stacked bar chart
    plt.figure(figsize=(8, 6))
    grouped_data.plot(kind='bar', stacked=True, figsize=(8, 6))
    plt.title(f'{col2} Distribution by {col1}')
    plt.xlabel(col1)
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.legend(title=col2)
    plt.tight_layout()
    plt.show()

# Interpretations
print("Interpretation of Stacked Bar Charts:")

print("\nSurvival Distribution by Pclass:")
print("The stacked bar chart shows the number of passengers who survived (1) and did not survive (0) within each passenger class (Pclass). We can observe the absolute counts of survivors and non-survivors for Pclass 1, 2, and 3. To understand survival *rates*, we would need to look at the proportion of survivors within each class.")

print("\nSurvival Distribution by Sex:")
print("This chart illustrates the number of male (likely 0) and female (likely 1) passengers who survived (1) and did not survive (0). It is evident that the number of male non-survivors is significantly higher than female non-survivors. We can also compare the number of survivors between the sexes.")

print("\nSurvival Distribution by Embarked:")
print("The stacked bar chart for Embarked and 2urvived shows the count of survivors (1) and non-survivors (0) for each embarkation point (0, 1, and 2). This allows us to see the absolute number of people from each embarkation point who survived or did not survive.")
```
